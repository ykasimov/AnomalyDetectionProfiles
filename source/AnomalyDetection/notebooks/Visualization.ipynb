```python
import json
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
```

```python
global_features = [
    'clientDestinationPortTotalBytesUDPEstablished',
    'clientDestinationPortNumberOfFlowsTCPEstablished',
    'clientDestinationPortNumberOfFlowsUDPNotEstablished',
    'clientDestinationPortTotalPacketsTCPEstablished',
    'clientDestinationPortNumberOfFlowsUDPEstablished',
    'clientDestinationPortTotalPacketsTCPNotEstablished',
    'clientDestinationPortTotalBytesUDPNotEstablished',
    'clientDestinationPortTotalBytesTCPEstablished',
    'clientDestinationPortTotalPacketsUDPNotEstablished',
    'clientDestinationPortNumberOfFlowsTCPNotEstablished',
    'clientDestinationPortTotalBytesTCPNotEstablished',
    'clientDestinationPortTotalPacketsUDPEstablished']
```

```python
profiles = 'CTU-Honeypot-Capture-2-2.json'
data = 'data'
with open(f'{data}/{profiles}','r') as f:
    data = json.load(f)

ips = {'CTU-Honeypot-Capture-1-4.json' : '192.168.2.16',
      'CTU-Honeypot-Capture-2-2.json' : '192.168.100.109',
       'CTU-Honeypot-Capture-2-1.json': '192.168.100.109',
       'CTU-Honeypot-Capture-3-1.json' : '192.168.100.64',
       'CTU-Honeypot-Capture-3-15.json' : '192.168.100.64',
       'CTU-Honeypot-Capture-2-6.json' : '192.168.100.109',
       'CTU-Honeypot-Capture-2-64.json' : '192.168.100.109',
       'CTU-IoT-Malware-Capture-2-2.profiles.json' : '192.168.100.103'}    

ip = ips[profiles]
```

```python
day= list(data[ip]['time'].keys())[0]
profiles = data[ip]['time'][day]

feature_name = global_features[0]

print(f'Feature: {feature_name}')

for key in profiles.keys():
    D = profiles[key][feature_name]
    print(key,D)
    figure(num=None, figsize=(20, 6), dpi=80, facecolor='w', edgecolor='k')
    plt.bar(range(len(D)), list([int(x) for x in D.values()]), align='center')
    plt.xticks(range(len(D)), list(D.keys()))
    plt.show()
```
